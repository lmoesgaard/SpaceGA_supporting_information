```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 8), width_ratios=[4, 4, 4])
names = {"spacega": "SpaceGA",
         "ga": "GA"
        }
palette = ["#80d380", "#1f90fe"]

for k, name in enumerate(["ga", "spacega"]):
    for i, protein in enumerate(["PARP1", "TRYB1", "UROK"]):
        ax = axes[k, i]
        ax.set_xlim(0,1)
        ax.set_ylim(0,8.5)
        for rep in range(1,4):
            subset = pd.read_parquet(f"../{name}/{protein}/{rep}/final.parquet")
            sns.kdeplot(subset.s, ax=ax, c=palette[k])
        ax.set_xlabel(f"Tanimoto similatity")
        ax.set_title(f"Top 1000 molecules from {names[name]} screen on {protein}")

fig.tight_layout(pad=2.0)
fig.text(0.01, 0.99, 'A)', fontsize=14, verticalalignment='top', horizontalalignment='left')
fig.text(0.01, 0.52, 'B)', fontsize=14, verticalalignment='top', horizontalalignment='left')
plt.savefig("../figures/similarity.pdf")
```
